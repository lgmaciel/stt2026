# Covariância

Mede *como* duas variáveis variam juntas.

- *positiva* - variáveis crescem juntas
- *negativa* - uma variável cresce à medida que a outra decresce

```python
import numpy as np

x = [10, 20, 30, 40, 50]
y = [5, 10, 15, 20, 25]

covariancia = np.cov(x, y)[0, 1]
print("Covariancia:", covariancia)
```

```python
import numpy as np

x = [10, 20, 30, 40, 50]
y = [50, 40, 35, 30, 25]

covariancia = np.cov(x, y)[0, 1]
print("Covariancia:", covariancia)
```
