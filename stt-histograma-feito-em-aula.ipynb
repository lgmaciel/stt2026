## Simulando dados

```python
import numpy as np
import pandas as pd

dados_alturas = np.random.normal(171, 8.5, 1000)
df_alturas = pd.DataFrame(data = dados_alturas)
df_alturas.hist(bins=15)
```

Pesos

```python
dados_pesos = np.random.normal(75, 17.2, 1000)

df_pesos = pd.DataFrame(data = dados_pesos)
df_pesos.hist(bins=15)
```
