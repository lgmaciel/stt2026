# Exercícios de regressão linear
## Preço da casa é determinado pelo tamanho?

Considere os dados abaixo:

|Tamanho (m²)| 	Preço (milhares de R$)|
|-|-|
|50| 	300|
|70| 	420|
|70|    350|
|90| 	540|
|120|   500|


Queremos encontrar a e b para que:

$\text{Preço} = a \times \text{Tamanho} + b$


Tarefa:

- Ajuste um modelo linear aos dados e faça a previsão do preço de uma casa de 75 m², 80 m² e 85 m².
