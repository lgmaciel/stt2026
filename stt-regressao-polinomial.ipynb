# Regressão polinomial

Vamos introduzir as regressões polinomiais estudando um modelo linear para ver suas limitações. Vamos usar dados simulados gerados aleatoriamente para ter maior controle sobre a distribuição e podermos gerar os casos problemáticos mais facilmente.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

Começamos gerando valores de `y` a partir de uma função linear. Vemos que os valores formam uma linha reta.

```python
X = np.random.rand(100,1) # 100 linhas, 1 coluna
y = 5 * X + 4  # uma função linear de X (ax + b)
plt.scatter(X, y)
```

### Adicionando ruído

Não é essa a distribuição que normalmente vemos nos dados. Eles não são tão "comportados". Para dar um pouco mais de veracidade aos dados, vamos introduzir *ruído*. 

Introduzimos ruído para simular interferência nos dados causados por fatores desconhecidos. Fazemos isso somando a cada um dos valores de `y` um valor aleatório. Para a interferência ser mais próxima de casos reais, vamos usar como ruído valores obtidos de uma distribuição normal (gaussiana).

```python
X = np.random.rand(100,1) # 100 linhas, 1 coluna
ruido = np.random.randn(100, 1) # (valores aleatórios com uma distribuição normal)

# uma função linear de X (ax + b) + ruído 
y = 5 * X + 4  + ruido
plt.scatter(X, y) 
```

#### Controlando o ruído

Agora vamos adicionar um controle sobre a amplitude do ruído, ou seja, para quanto (para mais ou para menos) o valor do ruído vai variar.

Fazemo isso multiplicando o valor do ruído (aleatório) por um outro valor (constante) que representará a amplitude.

```python
X = np.random.rand(100,1) # 100 linhas, 1 coluna
ruido = np.random.randn(100, 1) # (valores aleatórios com uma distribuição normal)
amplitude_ruido = 0.4

y = 5 * X + 4  + amplitude_ruido * ruido # uma função linear de X (ax + b) + ruído (valores aleatórios com uma distribuição normal)
plt.scatter(X, y)
```

Para ampliar o ruído, subimos o valor de `amplitude_ruido`

```python
X = np.random.rand(100,1) # 100 linhas, 1 coluna
ruido = np.random.randn(100, 1) # (valores aleatórios com uma distribuição normal)
amplitude_ruido = 1.2

y = 5 * X + 4  + amplitude_ruido * ruido # uma função linear de X (ax + b) + ruído (valores aleatórios com uma distribuição normal)
plt.scatter(X, y)
```

Agora conseguimos controlar bem melhor o formato da distribuição de valores e podemos simular melhor as situações que queremos estudar.

### Treinando o modelo de regressão linear com os dados

```python
regressao = LinearRegression()
regressao.fit(X, y)

print(f"a = {regressao.coef_}, b = {regressao.intercept_}")
```

### Fazendo previsões

```python
# usamos valores aleatórios para X

X_aleat = np.linspace(0, 1, 100).reshape(-1,1) # reshape(linhas, colunas): -1,1 transforma linhas em colunas (-1, "linhas automaticas")
X_aleat
```

```python
y_predito = regressao.predict(X_aleat)
y_predito
```

### Plotando a reta de regressão

```python
plt.scatter(X, y)
plt.plot(X_aleat, y_predito, color="r")
```

Os dados tinham origem em uma função linear. Eles só não formavam uma reta por causa do ruído. Essencialmente eram lineares. Por isso uma reta ajustou-se tão bem a eles.

## Regressão linear - programa completo

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

a = 5       # coef. angular
b = 4       # coef. linear
r = 0.4     # amplitude do ruído
n = 100     # quantidade de valores

X = np.random.rand(n,1) # 100 linhas, 1 coluna
y = 5 * X + 4  + r * np.random.randn(n, 1) # uma função linear de X (ax + b) + ruído (valores aleatórios com uma distribuição normal)
plt.scatter(X, y)


regressao = LinearRegression()
regressao.fit(X, y)

print(f"a = {regressao.coef_}, b = {regressao.intercept_}")


# usamos valores aleatórios para X

X_aleat = np.linspace(0, 1, n).reshape(-1,1) # reshape -1,1 transforma linhas em colunas

# prevemos os valores de y correspondentes as X_aleat gerados
y_predito = regressao.predict(X_aleat)

# plotamos a reta
plt.plot(X_aleat, y_predito, color="r")
```

## Regressão polinomial


No programa abaixo, substituímos apenas a função que gera os valores de y. A função agora não é linear. Ela contém um termo do 2° grau. É uma função que tem a forma:

$aX^2 + bX + c$

Vamos tentar ajustar um modelo linear ($aX + b$) a esses dados.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

a = -8       
b = 4
c = 3
n = 100     # quantidade de valores
r = 0.4     # amplitude do ruído

X = np.random.rand(n,1) # 100 linhas, 1 coluna

# uma função polinomial de X (ax^2 + bx + c) + ruído (valores aleatórios com uma distribuição normal)
y = a * X **2 + b * X  +c + r * np.random.randn(n, 1) 
plt.scatter(X, y)


regressao = LinearRegression()
regressao.fit(X, y)

print(f"a = {regressao.coef_}, b = {regressao.intercept_}")


# usamos valores aleatórios para X

X_aleat = np.linspace(0, 1, n).reshape(-1,1) # reshape -1,1 transforma linhas em colunas

# prevemos os valores de y correspondentes as X_aleat gerados
y_predito = regressao.predict(X_aleat)

# plotamos a reta
plt.plot(X_aleat, y_predito, color="r")
```

### Problema com o modelo linear

Neste exemplo, o algorítmo funcionou bem. Porém, estamos com um erro maior no início, meio e final da série.

Se avaliarmos o coeficiente de determinação veremos que ele deve ser baixo.

### Coeficiente de determinação R²

#### O que significa R²?

- R² = 0.72 → O modelo explica 72% da variação nos dados
- R² = 0.89 → O modelo explica 89% da variação
- R² = 1.00 → Ajuste perfeito (raramente acontece com dados reais)

```python
regressao.score(X, y)
```

### Ajustando modelo polinomial

Começamos repetindo os dados e a análise exploratória do exemplo anterior, de regressão linear.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

a = -8       
b = 4
c = 3
n = 100     # quantidade de valores
r = 0.4     # amplitude do ruído

X = np.random.rand(n,1) # 100 linhas, 1 coluna

# uma função polinomial de X (ax^2 + bx + c) + ruído (valores aleatórios com uma distribuição normal)
y = a * X **2 + b * X  +c + r * np.random.randn(n, 1) 
plt.scatter(X, y)
```

Para ajustar uma função polinomial aos dados, vamos preparar os dados de entrada do algorítimo (dados de treino). Este é um processamento de pré-treino.

### Pré-treino:

Elevaremos cada um dos dados aos graus do polinômio:

- se for de 2° grau, serão os graus 0, 1 e 2
- se for de 3°, serão os graus 0, 1, 2 e 3
- etc...

Esses novos dados serão os usados para treino do algorítmo de regressão linear.

```python
from sklearn.preprocessing import PolynomialFeatures

## pré-treino
p_features = PolynomialFeatures(degree=2)
X_p_features = p_features.fit_transform(X)
X_p_features
```

**PolynomialFeatures** é uma classe do scikit-learn que cria objetos que sabem transformar variáveis numéricas em novos termos polinomiais (features).

Isso significa que esses objetos criam termos como $x, x^2, x^3$, etc., a partir dos dados originais. Isso é necessário para modelos de regressão linear, pois é o que permite que o modelo capture relações não lineares entre as variáveis. 

Esses novos termos polinomiais permitirão ajustar um modelo de regressão polinomial que capture relações mais complexas nos dados que uma simples regressão linear.


O **degree** especifica o grau do polinômio a ser criado. Neste caso, `degree=2` significa que ele criará termos como $x, x^2$ mas não $x^3$.

### Treinamento do modelo polinomial

Agora usamos os valores polinomiais para treinar o modelo

```python
regressao = LinearRegression()
regressao.fit(X_p_features, y)

print(f"a = {regressao.coef_}, b = {regressao.intercept_}")
```

### Predições com o modelo polinomial

Como o modelo foi treinado usando PolinomialFeatures, precisamos usar PolinomialFeatures também para as predições.


```python
# usamos valores aleatórios para X
X_aleat = np.linspace(0, 1, n).reshape(-1,1) # reshape -1,1 transforma linhas em colunas

## vamos transformar os valores aleatórios de X (X_aleat)
X_aleat_polinomiais = p_features.transform(X_aleat) # transforma dados em PolinomialFeatures


# prevemos os valores de y correspondentes aos X_aleat_polinomiais 
y_predito = regressao.predict(X_aleat_polinomiais)

# plotamos o modelo
plt.plot(X_aleat, y_predito, color="r")
```

### Previsão de um ponto

Não temos as observações para os valores 0,25 e 0,75. Para fazer uma predição com esses valores, criamos um array contendo cada valor em um array (um array que contém arrays).

Como o modelo é polinomial, aplicamos PolinomialFeatures ao valores antes de passá-los ao método `predict()` do modelo.

```python
## vamos transformar os valores de X 
X_ponto = p_features.transform([[0.25], [0.75]])

# prevemos os valores de y correspondentes aos X_polinomiais 
y_predito = regressao.predict(X_ponto)
y_predito   # 
```

## Regressão polinomial - programa completo

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures

a = -8       
b = 4
c = 3
n = 100     # quantidade de valores
r = 0.4     # amplitude do ruído

X = np.random.rand(n,1) # 100 linhas, 1 coluna


## VALORES GERADOS POR UMA FUNÇÃO POLINOMIAL

# uma função polinomial de X (ax^2 + bx + c) + ruído (valores aleatórios com uma distribuição normal)
y = a * X **2 + b * X  +c + r * np.random.randn(n, 1) 
plt.scatter(X, y)


## PRÉ-TREINO

p_features = PolynomialFeatures(degree=2)
X_p_features = p_features.fit_transform(X)
X_p_features

## TREINAMENTO DO MODELO

regressao = LinearRegression()
regressao.fit(X_p_features, y)

print(f"a = {regressao.coef_}, b = {regressao.intercept_}")

## PREDIÇÕES COM O MODELO POLINOMIAL AJUSTADO

# usamos valores aleatórios para X (os mesmos usados para gerar os dados)
X_aleat = np.linspace(0, 1, n).reshape(-1,1) # reshape -1,1 transforma linhas em colunas

## vamos transformar os valores aleatórios de X (X_aleat)
X_aleat_polinomiais = p_features.transform(X_aleat) # transforma dados em PolinomialFeatures

# prevemos os valores de y correspondentes aos X_aleat_polinomiais 
y_predito = regressao.predict(X_aleat_polinomiais)


## PLOTAGEM DA CURVA POLINOMIAL AJUSTADA
# plotamos o modelo
plt.plot(X_aleat, y_predito, color="r")
```

### Determinamos a qualidade do ajuste (R²) para o modelo polinomial

Observe que passamos para o método `score()` os vales de X que foram transformados (no pré-treino) por PolinomialFeatures.

```python
regressao.score(X_p_features, y)
```
