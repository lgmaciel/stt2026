# Correlação


Mede o nível relacionamento entre duas variáveis - o quanto uma variável está relacionada a outra. Isso ajuda a fazer predições e a entender dependências entre variáveis.

*Coeficiente de correlação de Pearson* 

Usado para medir a relação linear entre duas variáveis: `valor entre -1 e 1`


Abaixo, há uma forte correlação entre `x` e `y`. Quando `x` varia, `y` varia junto de forma linear (no caso, $y=2x$)

```python
import numpy as np

x = [10, 20, 30, 40, 50]
y = [5, 10, 15, 20, 25]
correlacao = np.corrcoef(x, y)[0, 1]
print("Coeficiente de correlação de Pearson:", correlacao)
```

Abaixo, vemos que não há uma forte correlação entre as variáveis.

```python
import numpy as np

x = [1, 20, 3, 40, 5]
y = [5, 10, 15, 20, 25]
correlacao = np.corrcoef(x, y)[0, 1]
print("Coeficiente de correlação de Pearson:", correlacao)
```
