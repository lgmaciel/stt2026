# Mediana

Valor do meio de uma série de valores em ordem crescente.

Boa para dados com "outliers" porque não é afetada por valores muito discrepantes.

Para calcular:

$x = [n]$

$n = quantidade(x)$

Quando `n` é ímpar, a mediana é o valor que está na posição:

$pos = \frac{(n+1)}{2}$

$mediana = x[pos]$

Quando `n` é par, a mediana é a média dos dois valores que ocupam a posição do meio da série.



```python
import statistics

x = [5, 10, 20, 40, 50]

mediana = statistics.median(x)
print("Mediana:", mediana)
```
