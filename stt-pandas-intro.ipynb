### Criar uma Série

```python
# Série a partir de uma lista

datas_das_aulas = pd.Series([
    "20260319",
    "20260320",
    "20260323",
    "20260324",
    "20260325",
    "20260326",
    "20260330",
    "20260331",
    "20260401",
])

datas_das_aulas
```

```python
# Série a partir de uma tupla

sequencia = pd.Series((0,3,6,9,12))
sequencia
```

```python
# Série a partir de um dicionário

serie_notas = pd.Series(
    {
        "fulano": 7.5,
        "beltrano": 6.5,
        "ciclano": 5.0
    }
)
serie_notas

# note como as chaves viraram índices das linhas
```

```python
# Séria a partir de lista de valores e lista de índicies

serie_notas_versao2 = pd.Series(
    [7.5, 6.5, 5.0],
    index = ["fulano", "beltrano", "ciclano"]
    
)

serie_notas_versao2
```

### Pesquisa por intervalo - [início: fim]

```python
dataframe[:7] # do início ao ítem 7 (exclusive) - sete primeiros ítens
```

```python
dataframe[-7:] # do ítem 7 (inclusive) até o final
```

```python
dataframe[3:10] # do ítem 3 (quarta linha) até o ítem 10 (exclusive)
```

```python
dataframe[-5:] # os últimos 5 ítens - início = contando 5 de trás pra frente; fim: último ítem
```

```python
dataframe[:-5] # todos os ítens exceto os 5 últimos 
```

### Selecionando colunas por intervalo - iloc[linhas, colunas]

Este formato de pesquisa retorna um dataframe.

```python
dataframe.iloc[0:, 0:1] # retorna a primeira coluna
```

```python
dataframe.iloc[0:5, 0:1] # retorna as 5 primeiras linhas da primeira coluna
```

```python
dataframe.iloc[0:5, 1:2] # retorna as 5 primeiras linhas da segunda coluna
```

Este formato retorna uma série.

Note que aqui estamos passando apenas o índice da coluna e não um intervalo do tipo [início:fim]

```python
dataframe.iloc[0:5, 1] # retorna a primeira coluna
```

### Acessando os valores de uma coluna ou série em formato `numpy` - array

```python
dataframe.iloc[0:5, 1].values
```
