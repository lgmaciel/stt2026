# Desvio padrão

Assim como a variância, o desvio padrão também mede o quando os dados variam.

A diferença é que o valor do desvio padrão é dado na mesma unidade de medida que os dados.


Nos exemplos abaixo, se os valores da lista `x` representam "preços em Reais", então o valor do desvio padrão está em Reais também, e pode ser lido como "o quanto os preços variam de um para o outro" (considerando que eles estejam ordenados).

```python
import statistics

x = [35, 30, 32, 38, 40]
desviop = statistics.stdev(x)
print(f"Média: {statistics.mean(x)}")
print("Desvio padrão:", desviop)
```

```python
x = [5, 10, 20, 40, 50]
desviop = statistics.stdev(x)
print(f"Média: {statistics.mean(x)}")
print("Desvio padrão:", desviop)
```
