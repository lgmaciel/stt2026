Colocar os dados em rol

````python
alturas = [162, 168, 155, 170, 165, 162, 168, 160]
print(f"Rol: {sorted(alturas)}")
````

calcular a média

````python
from statistics import mean

mean(alturas)
````

criar uma tabela de frequências

Rol: [155, 160, 162, 162, 165, 168, 168, 170]

|valor | freq|

|160   |  1  |
|162   |  2  |

Tabela de frequência simples em python

````python
from collections import Counter
alturas = [170, 165, 168, 162, 168, 160, 162, 155]

tabela_freq = Counter(alturas)
print(tabela_freq)
print("ou, mantendo a ordenação do rol")
print(sorted(tabela_freq.items()))
````

Exibindo em formato de tabela

````python
print("Alt.| Freq.")
for altura, frequencia in sorted(tabela_freq.items()):
    print(f"{altura} | {frequencia}")
````

Quantos estão acima da média?

````python
alturas = [162, 168, 155, 170, 165, 162, 168, 160]
print(f"Rol: {sorted(alturas)}")
print(f"Média: {mean(alturas)}")
````

````python
def criterio(x):
    return x > 163.75

list(filter(criterio, alturas))
````

usando função anônima (lambda)

````python
list(filter(lambda x: x > 163.75, alturas))
````

Quem está abaixo da média?

````python
list(filter(lambda x: x < mean(alturas), alturas))
````

Quantos estão acima da média?

````python
len(list(filter(lambda x: x > mean(alturas), alturas)))
````

Quantos estão abaixo da média?

````python
len(
    list(
        filter(lambda x: x < mean(alturas), alturas)
    )
)
````
