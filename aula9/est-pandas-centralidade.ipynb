### Estatísticas básicas - describe()

Esta função calcula as estatísticas básicas para uma Série:

- contagem de observações (linhas)
- média (mean)
- desvio padrão (std)
- mínimo (min)
- quartis (1Q = 25%, 2Q = 50%, 3Q = 75%)
- máximo (max)

```python
serie.describe()
```
