## Medidas de dispersão

```python
import pandas as pd
import numpy as np
```

```python
temperaturas = pd.read_csv("dados/dia-mes-temp.csv")
temperaturas
```

### Amplitude

O valor máximo - o valor mínimo

```python
t = temperaturas["Temperatura °C"]
amplitude = t.max() - t.min()
amplitude
```

### Variância

```python
t.var()
```

### Desvio padrão

Raiz quandrada da variância

```python
np.sqrt(t.var())
```

```python
t.std()
```

```python
temperaturas["Temperatura °C"].std()
```

```python
consumos = pd.read_csv('dados/consumos-residencias.csv')
consumos
```

```python
consumos["consumo médio em kwh"].std()
```

```python
consumos.sort_values(
    by='consumo médio em kwh' # ordena os dados do dataframe por 'consumo médio em kwh
).plot(
    kind='bar',
    x='residências pesquisadas',
    y='consumo médio em kwh'
)
```

### Estatísticas básicas - describe()

Esta função lista, para cada coluna do dataframe, as estatísticas básicas:

- contagem de observações (linhas)
- média (mean)
- desvio padrão (std)
- mínimo (min)
- quartis (1Q = 25%, 2Q = 50%, 3Q = 75%)
- máximo (max)

```python
consumos.describe()
```
