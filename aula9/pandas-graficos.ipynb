# Gráficos com pandas


Plotar um gráfico de linhas usando todas as variávesi do dataframe - plot()

```python
import pandas as pd

df = pd.read_csv("dados/dia-mes-temp.csv")
df.plot()
```

Diagrama de linha para um par (x,y)

```python
import pandas as pd

df = pd.read_csv("dados/dia-mes-temp.csv")

df.plot(
    kind="line", # Definimos o tipo de gráfico
    x="Dia do Mês", # Valores do eixo x
    y="Temperatura °C", # Valores correspondentes de y
    color = "red"
    )
```

### Diagrama de dispersão - scatter plot

```python
import pandas as pd

df = pd.read_csv("dados/dia-mes-temp.csv")

df.plot(
    kind="scatter",
    x="Dia do Mês",
    y="Temperatura °C",
    color = "red"
    )
```

### Gráfico de barras



```python
df = pd.read_csv("dados/dia-mes-temp.csv")

df.plot(
    kind="bar",
    x="Dia do Mês",
    y="Temperatura °C",
)
```

### Histograma



Histogramas são adequados para séries com quantidades maiores de valores. Dividimos os valores em intervalos e plotamos o gráfico agrupando os valores nesses intervalos e plotando suas frequências (eixo y) acumuladas.

```python

import numpy as np
import pandas as pd

df = pd.DataFrame(
    data = np.random.randn(100, 1) # uma coluna com 100 valores aleatórios com distribuição normal - média 0 e desvio padrão 1
    )

df.plot(
    kind="hist", # tipo do gráfico
    bins=20, # número de barras - a quantidade de intervalos em que serão agrupados os dados
)
```
