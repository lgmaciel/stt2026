# Variância

Mede o quando os dados variam. É calculado tendo como referência a média.

Ajuda a entender a dispersão dos dados.

- variam muito? estão mais dispersos
- variam pouco? estão mais agrupados próximos da média

Variância alta: dados variam mais em relação à média

```python
import statistics

x = [5, 10, 20, 40, 50]
variancia = statistics.variance(x)
print(f"Média: {statistics.mean(x)}")
print("Variância:", variancia)
```

Variância baixa: dados estão mais próximos da média

```python
import statistics

x = [35, 30, 32, 38, 40]
variancia = statistics.variance(x)
print(f"Média: {statistics.mean(x)}")
print("Variânia:", variancia)
```
