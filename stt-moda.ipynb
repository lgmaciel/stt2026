# Moda

O valor mais frequente em uma série.

É mais útil aplicada a dados categorias ou nominais para identificar qual é a categoria mais comum.

```python
import statistics

x = [10, 20, 20, 30, 40]
c = ["alto","alto","médio","alto","baixo","baixo","médio","baixo","médio"]
moda_x = statistics.mode(x)
moda_c = statistics.mode(c)
print("Moda X:", moda_x)
print("Moda C:", moda_c)
```
