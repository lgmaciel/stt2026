## Estatísticas básicas com dados coletados em campo

Limpeza dos dados e conversão em list

```python

arq_pesos = open("dados/pesos.csv", "r", encoding="utf-8")
pesos_sujo = arq_pesos.readlines()
pesos = list(map(lambda x: int(x.strip()), pesos_sujo))
```

Algumas estatísticas com os dados

```python
print(f"Maior peso: {max(pesos)}")
```

```python
print(f"Menor peso: {min(pesos)}")
```

```python
print(f"Quantidade de observações: {len(pesos)}")
```

Média

```python
somatorio = sum(pesos)
quant = len(pesos)
media = somatorio/quant
media
```

usando função mean()

```python
from statistics import mean

print(f"Média dos pesos: {mean(pesos)}")
```
