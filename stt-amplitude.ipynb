# Amplitude

É a diferença entre o valor máximo e o mínimo de uma série.

Ajuda a enteder, de forma muito simples, a dispersão geral dos dos dados.

```python
x = [10, 20, 20, 30, 40]

amplitude = max(x) - min(x)
print(f"Amplitude: {amplitude}")
```

`Amplitude 30` significa que os dados estão dispersos "ao longo de" 30 unidades.
