# Regressão logistica

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
import matplotlib.pyplot as plt
```

## Altura + Peso => Academia sim ou não?

```python
dados = pd.read_csv("dados/IdadeAlturaPesoAcademia.csv",sep=";")
dados
```

```python
X = dados[['Peso', 'Altura']]
y = dados['Academia']
```

```python
# divide o dataset em dados de treino (70%) e teste (30%)
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=0)
```

```python
# instanciamos o modelo
log_regression = LogisticRegression()

# treinamos o modelo usando os dados de treino
log_regression.fit(X_train,y_train)
```

```python

# usamos o modelo para fazer predições usando os dados de teste
y_pred = log_regression.predict(X_test)
y_pred
```

diagnósticos

```python
# estas são as respostas que o modelo NÃO viu
y_test
```

```python
# Medimos o quanto as respostas do modelo se aproximam das respostas que ele não viu
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```
